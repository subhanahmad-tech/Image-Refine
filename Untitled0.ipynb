```python
# Install libraries (ONLY ONCE)
!pip install opencv-python-headless numpy
# Import libraries
import cv2
import numpy as np
from matplotlib import pyplot as plt
# Load image (MAKE SURE filename is correct)
img = cv2.imread('/content/Sample.jpeg')

# Convert BGR to RGB for display
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show original image
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')
plt.show()
# Brightness Enhancement
bright_img = cv2.convertScaleAbs(img, alpha=1, beta=50)

plt.imshow(bright_img)
plt.title("Brightened Image")
plt.axis('off')
plt.show()
# Contrast Enhancement
contrast_img = cv2.convertScaleAbs(img, alpha=1.5, beta=0)

plt.imshow(contrast_img)
plt.title("Contrast Enhanced Image")
plt.axis('off')
plt.show()
# Brightness + Contrast Combined
enhanced_img = cv2.convertScaleAbs(img, alpha=1.5, beta=50)

plt.imshow(enhanced_img)
plt.title("Final Enhanced Image")
plt.axis('off')
plt.show()
# Brightness + Contrast Combined
enhanced_img = cv2.convertScaleAbs(img, alpha=1.5, beta=50)

# Save Enhanced Image
cv2.imwrite('enhanced_image.jpg', cv2.cvtColor(enhanced_img, cv2.COLOR_RGB2BGR))
# Grayscale Image
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(gray_img, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')
plt.show()
```
